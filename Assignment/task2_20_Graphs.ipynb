```python

import pandas as pd
import matplotlib.pyplot as plt
import logging
```

```python
logging.basicConfig(level=logging.INFO)

logger = logging.getLogger("Mylogs")

gen_log_file = logging.FileHandler('dataAnalysisRecord.log')
gen_log_file.setLevel(logging.INFO)

logger.addHandler(gen_log_file)
```

Created Graph Using Matplotlib


```python
dfo = pd.read_excel("Superstore.xls")
logger.info("Dataset loaded successfully.")        
        
```

```python
# Histogram of sales distribution
def SalesDistribution(df):
    try:
        plt.figure(figsize=(8,5))
        df['Sales'].hist(bins=50, color='green', edgecolor='black')
        plt.xlabel('Sales')
        plt.ylabel('Frequency')
        plt.title('Sales Distribution')
        plt.show()
        logger.info("Sales distribution plotted successfully.")
    except Exception as e:
        logger.error(f"Error plotting sales distribution: {e}")
```

```python
SalesDistribution(dfo)
```

```python
# Scatter plot of Sales vs. Profit.
def SalesVsProfit(df):
    try:
        plt.figure(figsize=(8,5))
        plt.scatter(df['Sales'], df['Profit'], alpha=0.5, color='green')
        plt.xlabel('Sales')
        plt.ylabel('Profit')
        plt.title('Sales vs Profit')
        plt.show()
        logger.info("Sales vs Profit plotted successfully.")
    except Exception as e:
        logger.error(f"Error plotting Sales vs Profit: {e}")
```

```python
SalesVsProfit(dfo)
```

```python
#Bar chart of category-wise sales.
def categorySales(df):
    try:
        category_sales = df.groupby('Category')['Sales'].sum()
        category_sales.plot(kind='bar', color='yellow')
        plt.xlabel('Category')
        plt.ylabel('Total Sales')
        plt.title('Category-wise Sales')
        plt.show()
        logger.info("Category-wise sales plotted successfully.")
    except Exception as e:
        logger.error(f"Error plotting category sales: {e}")
```

```python
categorySales(dfo)
```

```python
# Bar chart of sub-category-wise sales.
def subcategorySales(df):
    try:
        subcat_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values()
        subcat_sales.plot(kind='barh', color='purple')
        plt.xlabel('Total Sales')
        plt.ylabel('Sub-Category')
        plt.title('Sub-Category-wise Sales')
        plt.show()
        logger.info("Sub-Category-wise sales plotted successfully.")
    except Exception as e:
        logger.error(f"Error plotting sub-category sales: {e}")
```

```python
subcategorySales(dfo)
```

```python
#Line chart for monthly sales trend.
def monthlySalesTrend(df):
    try:
        df['Order Date'] = pd.to_datetime(df['Order Date'])
        monthly_sales = df.resample('M', on='Order Date')['Sales'].sum()
        monthly_sales.plot(kind='line', marker='o', color='blue')
        plt.xlabel('Month')
        plt.ylabel('Total Sales')
        plt.title('Monthly Sales Trend')
        plt.show()
        logger.info("Monthly sales trend plotted successfully.")
    except Exception as e:
        logger.error(f"Error plotting monthly sales trend: {e}")
```

```python
monthlySalesTrend(dfo)
```

```python
# Pie chart of profit margin by category.
def profitMarginByCategory(df):
    try:
        category_profit = df.groupby('Category')['Profit'].sum()
        category_profit.plot(kind='pie', autopct='%1.1f%%')
        plt.title('Profit Margin by Category')
        plt.ylabel('')
        plt.show()
        logger.info("Profit margin by category plotted successfully.")
    except Exception as e:
        logger.error(f"Error plotting profit margin by category: {e}")
```

```python
profitMarginByCategory(dfo)
```

```python
# Bar chart of sales by region.
def RegionSales(df):
    try:
        region_sales = df.groupby('Region')['Sales'].sum()
        region_sales.plot(kind='bar', color='cyan')
        plt.xlabel('Region')
        plt.ylabel('Total Sales')
        plt.title('Region-wise Sales')
        plt.show()
        logger.info("Region-wise sales plotted successfully.")
    except Exception as e:
        logger.error(f"Error plotting region sales: {e}")
```

```python
RegionSales(dfo)
```

```python
# Scatter plot of Discount vs. Profit.
def discountVsProfit(df):
    try:
        plt.figure(figsize=(8,5))
        plt.scatter(df['Discount'], df['Profit'], alpha=0.5, color='magenta')
        plt.xlabel('Discount')
        plt.ylabel('Profit')
        plt.title('Discount vs Profit')
        plt.show()
        logger.info("Discount vs Profit plotted successfully.")
    except Exception as e:
        logger.error(f"Error plotting Discount vs Profit: {e}")
```

```python
discountVsProfit(dfo)
```

```python
# Box plot for sales distribution by state.
def salesByState(df):
    try:
        plt.figure(figsize=(10,6))
        states = df['State'].unique()
        data = [df[df['State'] == state]['Sales'] for state in states]
        plt.boxplot(data, vert=False, labels=states, patch_artist=True)
        plt.xlabel('Sales')
        plt.ylabel('State')
        plt.title('Sales Distribution by State')
        plt.show()
        logger.info("Sales distribution by state plotted successfully.")
    except Exception as e:
        logger.error(f"Error plotting sales distribution by state: {e}")
```

```python
salesByState(dfo)
```

```python
# Count plot for ship mode distribution.
def shipModeDistribution(df):
    try:
        plt.figure(figsize=(8,5))
        ship_mode_counts = df['Ship Mode'].value_counts()
        plt.bar(ship_mode_counts.index, ship_mode_counts.values, color='gray', alpha=0.7)
        plt.xlabel('Ship Mode')
        plt.ylabel('Count')
        plt.title('Ship Mode Distribution')
        plt.xticks(rotation=45)
        plt.show()
        logger.info("Ship mode distribution plotted successfully.")
    except Exception as e:
        logger.error(f"Error plotting ship mode distribution: {e}")
```

```python
shipModeDistribution(dfo)
```

Created Graph Using Plotly


```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import logging
```

```python
logging.basicConfig(level=logging.INFO)

logger = logging.getLogger("Mylogs")

gen_log_file = logging.FileHandler('plotlyDataAnalysisRecord.log')
gen_log_file.setLevel(logging.INFO)

logger.addHandler(gen_log_file)
```

```python
dfo = pd.read_excel("Superstore.xls")
logger.info("Dataset loaded successfully.")  
dfo.head(5) 
```

```python
# Line chart for order quantity over time
def orderQuantityTrend(df):
    try:
        df['Order Date'] = pd.to_datetime(df['Order Date'])
        monthly_orders = df.resample('M', on='Order Date')['Quantity'].sum()
        fig = px.line(monthly_orders, x=monthly_orders.index, y='Quantity', title='Order Quantity Trend')
        fig.show()
        logging.info("Order quantity trend plotted successfully.")
    except Exception as e:
        logging.error(f"Error plotting order quantity trend: {e}")

orderQuantityTrend(dfo)
```

```python
#Bar chart of average discount per state.
def averageDiscountPerState(df):
    try:
        avg_discount = df.groupby('State')['Discount'].mean().reset_index()
        fig = px.bar(avg_discount, x='State', y='Discount', title='Average Discount Per State')
        fig.show()
        logging.info("Average discount per state plotted successfully.")
    except Exception as e:
        logging.error(f"Error plotting average discount per state: {e}")

averageDiscountPerState(dfo)
```

```python
# Pie chart of sales contribution by customer segment.
def salesContributionBySegment(df):
    try:
        segment_sales = df.groupby('Segment')['Sales'].sum().reset_index()
        fig = px.pie(segment_sales, names='Segment', values='Sales', title='Sales Contribution by Segment')
        fig.show()
        logging.info("Sales contribution by segment plotted successfully.")
    except Exception as e:
        logging.error(f"Error plotting sales contribution by segment: {e}")


salesContributionBySegment(dfo)
```

```python
# Bubble chart of profit vs. quantity sold.
def profitVsQuantity(df):
    try:
        fig = px.scatter(df, x='Quantity', y='Profit', size='Sales', color='Category', title='Profit vs Quantity Sold')
        fig.show()
        logging.info("Profit vs quantity plotted successfully.")
    except Exception as e:
        logging.error(f"Error plotting profit vs quantity: {e}")

profitVsQuantity(dfo)
```

```python
# Line chart of sales trends by region.
def salesTrendByRegion(df):
    try:
        df['Order Date'] = pd.to_datetime(df['Order Date'])
        sales_region = df.groupby(['Order Date', 'Region'])['Sales'].sum().reset_index()
        fig = px.line(sales_region, x='Order Date', y='Sales', color='Region', title='Sales Trend by Region')
        fig.show()
        logging.info("Sales trend by region plotted successfully.")
    except Exception as e:
        logging.error(f"Error plotting sales trend by region: {e}")


salesTrendByRegion(dfo)
```

```python
# Bar chart of top 10 customers by profit.
def topCustomersByProfit(df):
    try:
        top_customers = df.groupby('Customer Name')['Profit'].sum().nlargest(10).reset_index()
        fig = px.bar(top_customers, x='Customer Name', y='Profit', title='Top 10 Customers by Profit')
        fig.show()
        logging.info("Top customers by profit plotted successfully.")
    except Exception as e:
        logging.error(f"Error plotting top customers by profit: {e}")


topCustomersByProfit(dfo)
```

```python
# Bar chart of products with negative profits.
def lossMakingProducts(df):
    try:
        loss_products = df[df['Profit'] < 0].groupby('Product Name')['Profit'].sum().nsmallest(10).reset_index()
        fig = px.bar(loss_products, x='Product Name', y='Profit', title='Top 10 Loss-Making Products')
        fig.show()
        logging.info("Loss-making products plotted successfully.")
    except Exception as e:
        logging.error(f"Error plotting loss-making products: {e}")

lossMakingProducts(dfo)
```

```python
# Histogram of sales distribution across cities.
def salesDistributionByCity(df):
    try:
        fig = px.histogram(df, x='City', y='Sales', title='Sales Distribution by City')
        fig.show()
        logging.info("Sales distribution by city plotted successfully.")
    except Exception as e:
        logging.error(f"Error plotting sales distribution by city: {e}")


salesDistributionByCity(dfo)
```

```python
# Heatmap of sales across sub-categories and regions.
def salesHeatMap(df):
    try:
        pivot_table = df.pivot_table(values='Sales', index='Sub-Category', columns='Region', aggfunc='sum')
        fig = px.imshow(pivot_table, title='Sales Heatmap Across Sub-Categories and Regions')
        fig.show()
        logging.info("Sales heatmap plotted successfully.")
    except Exception as e:
        logging.error(f"Error plotting sales heatmap: {e}")


salesHeatMap(dfo)
```

```python
# Q.26 Create a data frame where order date and ship date both are in JAN-2014

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] =  pd.to_datetime(df['Ship Date'])
jan_2014_df = df[
    (df['Order Date'].dt.month == 1) & (df['Order Date'].dt.year == 2014) &
    (df['Ship Date'].dt.month == 1) & (df['Ship Date'].dt.year == 2014)]

display("Data frame from Jan 2014 ", jan_2014_df)
```

```python
# Q.27 Find records having profit for consumer segment.

dfConsumerProfit = df[(df['Segment'] == 'Consumer') & (df['Profit'] > 0)]
dfConsumerProfit
```

```python
# Q.28 Generate new data frame for every state (hint: dynamic variable creation)

 #creating a unique column for state

superstoreUniqueState = df['State'].unique()

#storing dataframe in dict in each state

uniqueStateDfs = {}

for state in superstoreUniqueState:
    uniqueStateDfs[state] = df[df['State'] == state]


allStatesDf = pd.concat(uniqueStateDfs.values(), ignore_index=True)
print("Dispaly All state")
display(allStatesDf)


# pass the state name present in state col for display indivitual dataframe
display("Data frame for california")
cliforn = uniqueStateDfs['California']
```

```python
# Q.29 Generate the region wise,state wise average
# profit and sales and then extract information for east region.

# stateRegion
stateRegion = df.groupby(['Region', 'State'])[['Profit', 'Sales']].mean().reset_index()


# East Region information
eastRgionInfo = stateRegion[stateRegion['Region'] == 'East']
display(eastRgionInfo)

```
